```python
from sklearn.metrics import r2_score, mean_squared_error
def train_and_predict(model, features, target):
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print("mse score: %.2f" % mean_squared_error(y_test, y_pred))
    #(tn, fp, fn, tp) = confusion_matrix(y_test, y_pred).ravel()
    #print("confusion matrix")
    #print("tn, fp, fn, tp")
    #print(tn, fp, fn, tp)
    
naive_model = DecisionTreeRegressor(random_state=42)
param_model = DecisionTreeRegressor(max_depth=5, min_samples_split=10, min_samples_leaf=10, random_state=42)
train_and_predict(naive_model, features, target)
train_and_predict(param_model, features, target)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit(features)
scaled_features = scaler.transform(features)
model = SVR(C=1, kernel='rbf')
raveltarget=target.as_matrix().ravel()
train_and_predict(model, scaled_features, raveltarget)
```
